```python
from openai import OpenAI

client = OpenAI(
    # This is the default and can be omitted
    api_key='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',
)

prompts = [
    "Explain the concept of gravity in a simple way for an 8-year-old.",
    "What are some effective ways to teach basic algebra to students with learning disabilities?",
    "Create a short story about a space adventure, suitable for a child with autism.",
    "How can I make learning about historical events more engaging for students?",
    "Explain the water cycle to a student who loves video games.",
    "Design a simple experiment to demonstrate photosynthesis that a child can perform at home.",
    "What are some interactive online resources for learning geometry?",
    "How do you approach teaching reading to a student who is easily distracted?",
    "Create a motivational message for a student who is struggling with math.",
    "How can technology be used to improve learning outcomes for autistic students?",
    "Describe a game that can help with understanding fractions.",
    "What's a fun way to teach scientific method to young learners?",
    "Suggest a creative writing prompt for a high school student.",
    "How to explain internet safety to children in an engaging way?",
    "What are the benefits of using visual aids in teaching science?",
    "Give an example of how to use storytelling in math lessons.",
    "Recommend a mobile app that supports language learning for autistic users.",
    "How can parents support their child's learning at home with limited resources?",
    "What are some strategies for teaching empathy and social skills?",
    "Explain the importance of biodiversity in a way that resonates with young students."
]

# Function to query a model and print the response
def query_model(prompt, model):
  response = client.chat.completions.create(
      messages=[{"role": "user", "content": prompt,}],
      model=model,)
  print(f"Model: {model}\nPrompt: {prompt}\nResponse: {response.choices[0].message.content}\n\n")

# Evaluate each prompt on both models
for prompt in prompts:
    query_model(prompt, "ft:gpt-3.5-turbo-0125:personal:allyarc:99L5WQ0P")  # Fine-tuned model
    query_model(prompt, "gpt-3.5-turbo")  # Base model
```
